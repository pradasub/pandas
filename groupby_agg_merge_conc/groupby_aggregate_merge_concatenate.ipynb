```python
import pandas as pd
import numpy as np
```

# This is a pseudo data to practice merge concatinate in pandas

```python
# Students in my classes
# No duplicates in names
studentData = {
    'name' : ['jack', 'Riti', 'Aadi','prada','Ramesh','Niscal','zu','Eric'],
    'subject':['CS','Math','English','Physics','Physics','Stat','Math','CS'],
    'age' : [34, 30, 26,30,32,25,25,27],
    'city' : ['Sydney', 'Delhi', 'New york','ktm','ktm','Wilmington','Jakarta','Shanghai'],
    'salary':[8000, 1000,4000,10000,2500,3000,3000,10000]
}
```

```python
students = pd.DataFrame(studentData)
```

```python
# have duplicates whenever there is dual citizenship
countries = {
    'name' : ['jack','jack', 'Riti', 'Aadi','prada','Ramesh','Ramesh','tulsi'],
    'country':['US','UK','India','Canada','Nepal','Nepal','China','nz']
}
```

```python
students_countries = pd.DataFrame(countries)
```

```python
students
```

```python
students_countries
```

# Groupby

```python
# This directly groups by subject and calculates mean of numeric columns
students.groupby(['subject'],as_index=False).mean()
```

If you want subject to be an index

```python
students.groupby(['subject'],as_index=False).mean()
```

Groupby two fields

```python
students.groupby(['subject','city'],as_index=False).mean()
```

Only display the column wanted

```python
students.groupby(['subject','city'],as_index=False)['age'].mean()
```

# Aggregate

Aggregate over all, will display nan if we do not ask for it 

```python
students.aggregate({"salary":['sum','mean','min'],
                  "age":['min','mean']})
```

Aggregate over subject, calculate mean of age and sum of salaries, you have to rename the columns and reset the index

```python
students.groupby('subject').agg({'age':'mean', 
                                 'salary':'sum'}).rename(columns={'age':'avg_age', 
                                                                  'salary':'sum_salary'}).reset_index()
```

```python
students.groupby(['subject','city']).agg({'age':'mean', 
                                 'salary':'sum'}).rename(columns={'age':'avg_age', 
                                                                  'salary':'sum_salary'}).reset_index()
```

Repeat with count

```python
students.groupby(['subject']).agg({'age':'mean', 
                                 'city':'count'}).rename(columns={'age':'avg_age', 
                                                                  'city':'count_city'}).reset_index()
```

# Concatenate

lets concatenate two dataframe by columns

```python
# This will concatename by columns which is meaningless to us right now
pd.concat([students, students_countries], axis=1)
```

```python
# This will ust concatename by columns which is meaningless to us right now
pd.concat([students, students_countries], axis=0,sort=False)
```

# merge

```python
# The right side has duplicates and you can see how the duplicates are repeated
# So the duplicates on the key names are repeated
pd.merge(students,students_countries, on='name',how='inner')
```

```python
assert (len(students) >= len (pd.merge(students,students_countries, on='name',how='inner')))
```

# If you are trying to merge on primary key always check the duplicates 

```python
pd.merge(students,students_countries, on='name',how='left')
```

```python
# asset length on the left data is same as the left merge
```

```python
assert (len(students) == len(pd.merge(students,students_countries, on='name',how='left')))
```

Assertion error is due to the fact that there are duplicates in the data

Trying to merge left, if there are duplicates on the data on right then those will be repeated, hence we have to carefully examine
